```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

readfile = '人口統計.csv'
df = pd.read_csv(readfile,encoding="shift-jis",na_filter=False)
df = df.iloc[::-1]
sns.set(style = 'whitegrid')
fig, ax = plt.subplots(figsize = (7, 5))
plt.subplots_adjust(left = 0.2, right = 0.85, bottom = 0.05, top = 1.0)
df["man"] *= -1
age_colors = ["#4169e1", "#ff1493"]
age_names = ["man","woman"]
for name, color in zip(age_names, age_colors):
    sns.barplot(x = name, y = df["age"], data = df, color = color, label = name,
    orient = 'h', order = df["age"], ax = ax)
    yticklabels = ["100-","95-99","90-94","85-89","80-84","75-79","70-74","65-69","60-64","55-59","50-54",
                   "45-49","40-44","35-39","30-34","25-29","20-24","15-19","10-14","5-9","0-4"]

ax.set_xlabel("Number of people (unit: 10,000 people)")
ax.set_ylabel("age", fontsize = 12)
ax.set_xlim(-600, 600)
ax.set_xticklabels(['600', '400', '200', '0', '200', '400', '600'])
ax.set_yticklabels(yticklabels)
ax.legend(loc = 'lower left')
plt.savefig('graph.png')
plt.show()
plt.close()

```
